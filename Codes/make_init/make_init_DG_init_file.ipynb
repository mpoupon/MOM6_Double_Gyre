# Create initialization files following double-gyre spin up
Code creates initialization files with vertical profiles of temperature, salinity and COBALT tracers, using output from the spin up. Note that the higher resolution simulations will have grid-points between the edge of the domain and the first ocean grid-point of the lower resolution simulations. In the initialization files, the output of the spin up is thus extended past the last ocean grid-points by repeating the values at the last ocean grid-points beyond the edges of the domain.

Although most tracers in the spin up were initialized from the global model output, six tracers ('alk', 'dic', 'no3', 'o2', 'po4', 'sio4') were initialized from observational means and had units that differed from the `mol/kg` model units. The `field_table` for the spin up specified the observational units. However, in simulations initialized from the spin up fields, all units in `field_table` should be specified to `none` to indicate that the units of the initialization files match those of the model. See the source file `generic_tracer_utils.F90` for all allowed cases and conversions. 

### Script output
- `DG_spinup_Y[Ryr]_[fields]_*deg.nc`: initialization profiles for temperature, salinity and COBALT fields, using annual mean from spin up. `[Ryr]` indicates the year restarted with higher vertical resolution, and `[fields]` indicates the fields included (`TS` for physics only, `TStracers` for physics + COBALT).

### Steps
- Run one more spin-up year with a higher vertical resolution (75 layers instead of 35), using a vertical axis created with `DG_MakeHighResVertCoord.ipynb` and `diag_table_SpinUpInitFields` which will create a file (`hist_control_init3d_yearly__[Ryr].nc`) with all variables that need initializing.
- Build the initialization files at 1$^\circ$-resolution, letting the model do the best interpolation at the start of subsequent simulations.

```python
import os
import re
import cftime
import numpy as np
import xarray as xr
import netCDF4 as nc
from tqdm import tqdm
from scipy import interpolate
import matplotlib.pyplot as plt
```

```python
### Manual input ###
# Flags
fcheck  = 1  # plot a figure with the last variable to check it worked (1=yes, 0=no)
ftrac   = 1  # include COBALT tracers (1=yes, 0=no)
fordvm  = 1
fromdvm = 0

# Set fill value
fillval = -1e+20  # (Raphael/Enhui: -1e+20; Elizabeth:-1.e+34)

# Framework is in place to customize interpolation resolution, but trying to use
# 1 degree since this is the only information the spin up provides
hres = 1/9 # 1, 1/9, 1/27
hres_str = 'R85'
if fordvm:
    hres_str = hres_str + '_DVM'

# Note: the interpolation is not scaled by distance in km because 1. we do not 
# currently interpolate and 2. the model will need to equilibrate so probably minor

# String identifying what fields are included in the initialization file
if ftrac:
    strSU = 'TStracers'  # 'TS' for physics only; 'TStracers' for phys + COBALT
else:
    strSU = 'TS'
    
# Paths 
spath = '../../init/R9_spinupwithDVM_10yrs/'  # to higher resolution output
ipath = '../../init/R9_spinupwithDVM_10yrs/'  # save initialization file in INPUT folder

# Restart year with higher resolution 
Ryr   = '2000'

# Find simulation name (assuming the notebook is ran in ./INPUT/code/)
sim   = '9km_SpinUp' #re.findall(r"([A-Za-z+\d_]*)/INPUT", os.getcwd())

# Output filename
fname = 'DG_init_file'+hres_str+'.nc'

# Boundaries for 1-degree initialization files
latG1 = 15.5
latG2 = 69.5
lonG1 = 300
lonG2 = 350

# List all initialized variables 
# Physics 
varphys  = ['temp', 'salt']

varzero   = ['allzero']

# COBALT
# Use list that was previously initialized from global model output
tracers  = ['cadet_arag', 'cadet_calc', 'cased', 'chl', 'co3_ion', 'fed', 
            'fedi', 'felg', 'fesm', 'htotal', 'irr_mem', 'ldon', 'ldop', 'lith', 
            'nh4', 'nbact', 'ndi', 'nlg', 'nsm', 'nlgz', 'nmdz', 'nsmz', 
            'fedet', 'lithdet', 'ndet', 'pdet', 'sidet', 'silg', 'sldon', 
            'sldop', 'srdon', 'srdop', 'cadet_arag_btf', 'cadet_calc_btf', 
            'lithdet_btf', 'ndet_btf', 'pdet_btf', 'sidet_btf', 'mu_mem_ndi', 
            'mu_mem_nlg', 'mu_mem_nsm', 'fedet_btf', 'nh3']  

if fromdvm:
    tracers += ['nvmmdz','nvmlgz','nvmmdz_gut','nvmlgz_gut','nvmmdz_met','nvmlgz_met']

# Add list that was previously initialized from observations
# TAlk (alk in output), TCO2 (dic), NO3 (no3), O2 (o2), PO4 (po4), SiO4 (sio4)
tracers  = ['alk', 'dic', 'no3', 'o2', 'po4', 'sio4'] + tracers + varzero
```

```python
### Set resolution ###
dx = hres  
dy = hres

### Create longitude, latitude, and time vectors ###
longitude = np.arange(lonG1, lonG2, dx)
longitude = np.resize(longitude, (len(longitude), 1))
latitude  = np.arange(latG1, latG2, dy)
latitude  = np.resize(latitude, (len(latitude), 1))
Time      = 1
```

```python
### Loop through appropriate initialization variables ###
# If tracer included
if ftrac:
    # Add all variables
    var0     = varphys + tracers
else:
    # Use only T/S 
    var0     = varphys
```

# T, S, and tracers

```python
# Load high-resolution output files with all initialized fields
#dsH_ts = xr.open_dataset(ipath+'hist_endspinup_dynamics3d_yearly__2025_12_31.nc', decode_times=False).isel(time = slice(-2,-1)).load() # last day
#dsH_tr = xr.open_dataset(ipath+'hist_endspinup_cobalt3d_yearly__2025_12_31.nc', decode_times=False).isel(time = slice(-2,-1)).load()   # last day

#dsH_ts = xr.open_dataset(ipath+'hist_endspinup_dynamics3d_yearly__2025_12_31.nc', decode_times=False).load() # last day
#dsH_tr = xr.open_dataset(ipath+'hist_endspinup_cobalt3d_yearly__2025_12_31.nc', decode_times=False).load()   # last day

dsH_ts = xr.open_dataset(ipath+'hist_endspinup_dynamics3d_yearly__2009_12_31.nc', decode_times=False).load() # last day
dsH_tr = xr.open_dataset(ipath+'hist_endspinup_cobalt3d_yearly__2009_12_31.nc', decode_times=False).load()   # last day

dsH = dsH_ts
```

```python
## Create new vectors for expansion and interpolation ###
# New depth vector - add depth = 0 and depth = 4000 m to z_l
depth0 = np.append(0, dsH.z_l.values)
depth0 = np.append(depth0, 4000)

# Horizontal resolution of matrix
dx_in = np.mean(np.diff(dsH.xh))
dy_in = np.mean(np.diff(dsH.yh))

# New x vector - add longitude to East and West past limits for future interpolation
xhW = np.arange(lonG1 - dx_in/2, np.min(dsH.xh.values)+360, dx_in)  # longitude added to the East
xhE = np.arange(np.max(dsH.xh.values)+360 + dx_in, lonG2 + 3*dx_in/2, dx_in)  # longitude added to the West
xh0 = np.append(xhW, dsH.xh+360)
xh0 = np.append(xh0, xhE)

# New y vector - add latitude to North and South past limits for future interpolation
yhS = np.arange(latG1 - dy_in/2, np.min(dsH.yh.values), dy_in)  # latitude added to the South
yhN = np.arange(np.max(dsH.yh.values) + dy_in, latG2 + dy_in/2, dy_in)  # latitude added to the North
yh0 = np.append(yhS, dsH.yh)
yh0 = np.append(yh0, yhN)
```

```python
### Start initialization file to be saved ###
writing = nc.Dataset(spath+fname,"w",format="NETCDF3_64BIT_OFFSET")
time    = writing.createDimension("time", 1)
depth   = writing.createDimension("depth", len(depth0))
lat     = writing.createDimension("lat", len(latitude))
lon     = writing.createDimension("lon", len(longitude))

lat     = writing.createVariable("lat","f8",("lat"))
lat[:]  = latitude
lat.standard_name    = "latitude"
lat.units            = "degrees_north"
lat.cartesian_axis   = "Y"
 
lon     = writing.createVariable("lon","f8",("lon"))
lon[:]  = longitude
lon.standard_name    = "longitude"
lon.units            = "degrees_east"
lon.cartesian_axis   = "X" ;

time    = writing.createVariable("time","f8",("time"))
time[:] = 0.0
time.standard_name   = "time"
time.units           = "days since 0001-01-01 00:00:00"
time.calendar        = "noleap" ;
time.cartesian_axis  = "T" ;
    
depth    = writing.createVariable("depth","f8",("depth"))
depth[:] = depth0
depth.standard_name  = "depth"
depth.units          = "m"
depth.positive       = "down"
depth.cartesian_axis = "Z" ;
```

```python
### Loop through all the variables ###
for var in tqdm(var0):
    
    if var in varphys:
        dsH = dsH_ts
        v_in = dsH[var].isel(time=-1).values     # (time,) depth, latitude, longitude
    
    elif var in varzero:
        dsH = dsH_tr
        v_in = dsH['nlgz'].isel(time=-1).values * 0 # (time,) depth, latitude, longitude
    
    elif (fromdvm and (var == 'nvmmdz_met')):
        dsH = dsH_tr
        v_in = dsH['nmvmdz_met'].isel(time=-1).values
        
    elif (fordvm and (var in ['nlgz', 'nmdz'])):
        dsH = dsH_tr
        v_in = dsH[var].isel(time=-1).values /2 # (time,) depth, latitude, longitude

    else:
        dsH = dsH_tr
        v_in = dsH[var].isel(time=-1).values     # (time,) depth, latitude, longitude


    # Repeat the last row of matrix - top and bottom
    v_exp = np.concatenate([np.expand_dims(v_in[0,:,:], axis=0), v_in], axis=0)  # Append at top
    v_exp = np.concatenate([v_exp, np.expand_dims(v_in[-1,:,:], axis=0)], axis=0)  # Append at bottom

    # Repeat the row at the East and West
    v_W = np.tile(np.expand_dims(v_exp[:,:,1], axis=2), (1,1,len(xhW)+1))  # repeat 6 times (depth, latitude, longitude)
    v_E = np.tile(np.expand_dims(v_exp[:,:,-2], axis=2), (1,1,len(xhE)+1))  # repeat 6 times (longitude, depth, latitude)
        # Add + 1 to overwrite NaN rows

    # Append - East and West
    v_exp = np.concatenate([v_W, v_exp[:,:,1:-1]], axis=2)  # exclude rows with NaN
    v_exp = np.concatenate([v_exp, v_E], axis=2)

    # Repeat the row at the South and North
    v_S = np.tile(np.expand_dims(v_exp[:,1,:], axis=1), (1,len(yhS)+1,1))  # repeat 3 times (depth, latitude, longitude)
    v_N = np.tile(np.expand_dims(v_exp[:,-2,:], axis=1), (1,len(yhN)+1,1))  # repeat 4 times (depth, latitude, longitude)
        # Add + 1 to overwrite NaN rows

    # Append - North and South
    v_exp = np.concatenate([v_S, v_exp[:,1:-1,:]], axis=1)  # exclude rows with NaN
    v_exp = np.concatenate([v_exp, v_N], axis=1)

    # Create empty matrix to receive interpolation
    v_out = np.full((len(depth0), len(latitude), len(longitude)), np.NaN)  # depth, latitude, longitude

    # Interpolate to domain edges 
    for idepth in np.arange(0, len(depth0)):

        f1 = interpolate.interp2d(xh0, yh0, np.squeeze(v_exp[idepth,:,:]), kind='cubic')  # depth, latitude, longitude
        v_out[idepth,:,:] = f1(longitude.flatten(), latitude.flatten())

    # Expand dimensions for time
    v_out = np.expand_dims(v_out, axis=0)  # time, depth, latitude, longitude
    
    # Add variable to saving structure
    temp           = writing.createVariable(var,"f4",("time","depth","lat","lon"), fill_value=fillval)
    temp[:,:,:,:]  = v_out
    
    
    if var in varzero:
        temp.long_name     = 'Zero initialisation for concentration'
        temp.units         = dsH['nlgz'].units
    
    elif var == 'nvmmdz_met':
        temp.long_name     = dsH['nmvmdz_met'].long_name
        temp.units         = dsH['nmvmdz_met'].units
    
    else:
        temp.long_name     = dsH[var].long_name
        temp.units         = dsH[var].units
    temp.description   = "from " + sim[0]

# Write the file
writing.close()
```

```python
### Check that all looks fine (use only during troubleshooting; could be put in loop) ###
if fcheck:
    fig, axs = plt.subplots(1, 2, figsize=(15, 5), sharey=False)
    axs = axs.flatten()

    vmin = np.min(v_exp[1,:,:])*0.8
    vmax = np.max(v_exp[1,:,:])

    xlim = (np.min(xh0), np.max(xh0)-1) 
    ylim = (np.min(yh0), np.max(yh0)) 

    # Input
    pT = axs[0].pcolormesh(dsH.xh+360, dsH.yh, v_in[1,:,:], cmap='inferno', vmin=vmin, vmax=vmax)
    fig.colorbar(pT, ax=axs[0])
    axs[0].set_ylabel('Latitude')
    axs[0].set_xlabel('Longitude')
    axs[0].set_title('Spin up output')

    axs[0].set_xlim(xlim)
    axs[0].set_ylim(ylim)

    # Output
    pE = axs[1].pcolormesh(longitude.flatten(), latitude.flatten(), v_out[0,1,:,:], 
                           cmap='inferno', vmin=vmin, vmax=vmax)
    fig.colorbar(pT, ax=axs[1])
    axs[1].set_ylabel('Latitude')
    axs[1].set_xlabel('Longitude')
    axs[1].set_title('Expanded')

    axs[1].set_xlim(xlim)
    axs[1].set_ylim(ylim)
```
